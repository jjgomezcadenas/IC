DNNcomparison just gives a plot of error as a function of epoch for all the saved histories in a .h file

It's purpose is essentially to compare the performace of different size neural networks in learning the x,y coords for a set number of EL hits.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy  as np
import tables as tb


N_ELpts = 1
show_1_layer = False # change to True to show error plot for single layer network
                     # default set to false because this makes plot ugly
```

Load the file containing the history corresponding to N_ELpts

```python
#hist_filename = 'drophist_' + str(N_ELpts) + 'hit.h'  # load file with some dropout
#hist_filename = 'adamhist_' + str(N_ELpts) + 'hit.h'      # load file with some dropout
hist_filename = 'hist_' + str(N_ELpts) + 'hit.h'
histories = []
nodenames = []
f = tb.open_file(hist_filename, 'r+')
for node in f.root:
    histories.append(node[0])       # get error arrays
    nodenames.append(node._v_name)  # get error array names
histories = np.array(histories)
f.close()
```

Now plot the error:

```python
start = 0

for i in range(start,len(nodenames)): 
    plt.plot(range(0,histories.shape[1]),histories[i],label=nodenames[i])
plt.legend(loc='upper right')
plt.grid(True)
plt.xlabel('Epochs')
plt.ylabel('Mean error per hit (mm)')
plt.title(hist_filename)
plt.show()
```

```python
start = 1

for i in range(start,len(nodenames)): 
    plt.plot(range(0,histories.shape[1]),histories[i],label=nodenames[i])
plt.legend(loc='upper right')
plt.grid(True)
plt.xlabel('Epochs')
plt.ylabel('Mean error per hit (mm)')
plt.title(hist_filename)
plt.show()
```
